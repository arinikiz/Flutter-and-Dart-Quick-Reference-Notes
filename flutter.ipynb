Every screen in Flutter is ultimately built as a widget tree:
- Eaach widget tells Flutter how it wants to be drawn and how it should layout its children.
- The framework recursively builds this widget tree.

```sql

MaterialApp
 └── Scaffold
      ├── AppBar
      └── Body (like Column, Row, etc.)

```

Widgets in Flutter can be categorized into two:
1. Stateles Widgets
2. Stateful Widgets


A **Stateless Widget** is immutable.
- Meaning that ince it's built, its configuration (properties) cannot change during the widget's lifetime unless the entir widget is replacced by its parent in the Widget Tree.
- Stateless widgets are ideal for static element whose appearance depends solely on the configuration data passed to them via their constructor- i.e., ideal for UI that doesn't change with user interaction.
```raw

```
A **Stateful Widget** is dynammic.
- Meaning that it can change appearance in response to event tirggered by user interaction, network data, or system events.
- This dynamism is managed by seperating the widget definition from its mutable state.
    - A stateful widget delegates responsibility for managinng mutable data to a seperate object called the State object.
- Stateful widgets are essential for interactive elements.
```raw




Widgets are also classified in conceptual, organizational grouping which is used by developers and documentation to create widgets based on their design purpose and role in the UI hierachy:

1. Core/Atomic Widgets:
- These are the most basic building blocks of any Flutter UI.
- They typically render a fundamental piece of content or an interactive element.
- Examples: `Text`, `Image`, `Icon`, `ElevatedButton`

```raw

```

**`Column`**

It is a multi-child layout widget in Flutter that arranges its children vertically, **one below another (i.e., stacked)**, along the main axis.  

It is the vertical counterpart of the `Row`, that is **widgets are stacked on top of each other**

```dart

Column({
  MainAxisAlignment mainAxisAlignment = MainAxisAlignment.start,
  CrossAxisAlignment crossAxisAlignment = CrossAxisAlignment.center,
  List<Widget> children = const <Widget>[],
})

```
- `mainAxisAlignment` --> Controls how children are placed vertically.
  - Common values are:
    - `MainAxisAlignment.start` --> (default): Children are at the top
    - `MainAxisAlignment.center` --> Children are centered vertically
    - `MainAxisAlignment.end` --> Children are at the bottom
    - `MainAxisAlignment.spaceBetween`--> First at the top, last at the bottom, equal gaps between others
    - `MainAxisAlignment.spaceAround` --> Equal space around each child (half at edges)
    - `MainAxisAlignment.spaceEvenly` --> Equal space everywhere, including top and bottom.  

- `crossAxisAlignment` --> Controls horizontal alignment of children.
  - Common values are:
    - `CrossAxisAlignment.center` --> center horizontally (default)
    - `CrossAxisAlignment.start` --> align to the left
    - `CrossAxisAlignment.end` --> align to the right
    - `CrossAxisAlignment.stretch` --> expand each child to fill width  


- `children` --> The list of widgets you want inside the column.





**`Row`**

Row is a layout widget that arranges its children horizontally- from left to right (or right to left depending on text direction)  

It's the horizontal version of `column`, that is **that is the widgets are stacked side by side**  

The barebone general sytnax of `Row` is:
``` dart

Row({
  List<Widget> children = const <Widget>[],
  MainAxisAlignment mainAxisAlignment = MainAxisAlignment.start,
  CrossAxisAlignment crossAxisAlignment = CrossAxisAlignment.center,
})

```

- `mainAxisAlignment` --> Controls how the children are placed horizontally
    - Type is `MainAxisAlignment`
    - The common values are:
        - `MainAxisAlignment.start` --> children aligned to the left (default)
        - `MainAxisAlignment.center` --> children centered horizontally
        - `MainAxisAlignment.end` --> children aligned to the right
        - `MainAxisAlignment.spaceBetween` --> Places first child at start, the last at end, evenly spaces the rest (no space before the first child or after the last child)
        - `MainAxisAlignment.spaceAround` --> Equal space around each child (half space at edges, full spaces in between)
        - `MainAxisAlignment.spaceEvenly` --> Equal space between and around all children

- `crossAxisAlignment` --> controls vertical alignment
    - Type is `CrossAxisAlignment`
    - Most common values are:
        - `CrossAxisAlignment.start` --> Align children to the top
        - `CrossAxisAlignment.center` --> Centers children vertically 
        - `CrossAxisAlignment.end` --> Aligns children to the bottom


- `children` --> The list of widgets to align (i.e., stack side by side)


```raw


```

```dart

import 'package:flutter/material.dart';

void main() {
  runApp(const MyApp());  // Root widget
}

class MyApp extends StatelessWidget {
  
  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      title: 'Flutter Demo',
      home: const HomeScreen(),
    );
  }
}

## Coding Stateful Widgets

In Flutter, state refers to the data that can change over time within your app.  

This includes things like the current value of a counter, iser imput in a text fields, or the result of asynchromous operation.  

The state determines what the UI looks like at any given moment.  

When the state changes, Flutter rebuilds pars of the widget treee to reflect the new data.  

We can formally define state as:
> The mutable information stored within the `State` object that determines a widget's appearance and behavior.  
When this data chnages, the Flutter framework automatically triggers a rebuild of the affected widgets, ensuring that the UI reflects the current state.  


Flutter widgets are **immutable**, meaning that their configuration cannot change once created.  

To make the UI dynamic, you seperate:
- **Widget (immutable)** --> Describes how something should look
- **State (mutable)** --> Holds what it currently is (data that can change)  

The **State object** acts as the memory of a widget. When you change something inside the state, Flutter calls the `setState()` method, which rebuilds the UI with updated information.  

In this regard a **Stateful Widget** is a widget that has the `State` object.  

Note that StatefulWidget itself is just a blueprint that describes what the UI should look like. It in itself is not mutable. So you have to manually turn this blueprint into a mutable reality.

To do so:
1. Extend `Stateful Widget`
    - > `class yourWidgetName extends StatefulWidget`  
        - This declares a widget that can have changing data. The widget class itself stays immutable; it only defines identity/config  
2. Add a Constructor that forwards keu
    - > `const StateDemoScreen({super.key})`  
        - Altough you rarely touch the `key` yourself, it helps Flutter track this widget in the tree duting rebuilds.
        - `const` enables compile-time instatnitation and minor optimizations.  

3. Override `createState()`
    - This is the required factory method the framework calls to create the **mutable companion object (i.e., the `State`)**  
    - By overriding it, you tell Flutter which `State` class to attach to this widget.  

4. Define the private `State` class
    - Anything that can change wile the app runs lives in this `State` class.  

    - You also override `build()` within this class


```raw

```
So your main structure will look like:
```dart
class StateDemoScreen extends StatefulWidget {

	const StateDemoScreen({super.key});
	
	@override
	state<StateDemoScreen> createState() => _StateDemoScreenState();
	
}


class _StateDemoScreenState extends State<StateDemoScreen> {
	
	// Mutable data and your build() lives here
}




Note that when overriding `build()` for Stateful widgets, you can return `MaterialApp`, BUT you should not do so (except in very specific cases).  

The `MaterialApp` is the root widget of most Flutter apps. It sets up:
- Global app-level settings
- The app-wide `Navigator` and `ScaffoldMessenger`
- Default material design behavior  

So usually, the `MaterialApp` lives only once, right in your `main()`.  

Returning the `MaterialApp` in your StetfulWidget extension means that everytime you call `setState()`, `MaterialApp` rebuilds.
- Rebuilding a `MaterialApp` means reinitializing routing, themes, navigation stack- all expensive operations.  

Hence:
> The root `MaterialApp` should be devlared once and your screen widgets (Stateful or Stateless) shoudl return Scaffold or lower-level layout widgets instead.  

**Return the `MaterialApp` from the top-level app widget's `build()`- the one you pass to `runApp(...)`**



```raw

```

You never call `createState()` yourself, you just override it.
- `createState()` is a framework-level hook that Flutter itself calls excatlly once, when it first inserts your `StatefulWidget` into the widget tree.  
```raw

```
On the other hand, you call `setState()` manually when your state's data changes.
- It's like telling Flutter something changed- please rebuild my UI.  

You use `setState()` by:
1. Calling it in your change function
2. And you define your change operation as an anonymous callback function for the `setState()` function's parameter:  

```dart
void yourChangeFunction(){
    setState(() {
    // your mutable variables change here
    });
}
```

```raw

```
Full exmaple:
```dart
class _CounterScreenState extends State<CounterScreen> {
  int _count = 0; // state variable

  void _increment() {
    setState(() {
      _count++; // change the state here
    });
  }

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      body: Center(child: Text('Count: $_count')),
      floatingActionButton: FloatingActionButton(
        onPressed: _increment, // triggers setState()
        child: const Icon(Icons.add),
      ),
    );
  }
}
```

Notice that we are making our change function, as well as the State class of the widget privete (`_`):
- This is so that no other code outside the file should directly mess with how this screen's state is managed.
- This keeps your widget self-contained and avoids accidental misuse.

```raw


```

Now you know everyhting, here is an implementation example for a simple counter app:

```dart
class DynamicCounter extends StatefulWidget {

  const DynamicCounter({super.key});

  @override
  State<DynamicCounter> createState() => _StateDynamicCounter();
}

class _StateDynamicCounter extends State<DynamicCounter> {

  int _count = 0;

  _increment() {
    setState(
        (){
          _count++;
        }
    );
  }

  @override
  Widget build(BuildContext context){
    return Scaffold(
      appBar: AppBar(
        title : Text("CounterApp")
      ),
      floatingActionButton: FloatingActionButton(onPressed: _increment, // <== Note that you have to pass a function reference. If you pass _increment(), the functionality won't work as intended
      child: Text("+")),
      body: Center(child: Text("Counter: $_count"),),
    );
  }

}
```

## Widget Navigation

In a aoo with multiple screens, each unique screen is typically its own top level widget.  

For instance, each class that you extend as widget is a screen:
```dart
class Thrivetrack extends StatelessWidget { // <== This would represent a single screen
 
  @override
  Widget build(BuildContext context){
    return MaterialApp(
      home : Scaffold(
        appBar: AppBar(
          title: Text("ThriveTrack"),
          backgroundColor: Colors.blueAccent
        ),
        body : Center(child: Column(
          mainAxisAlignment: MainAxisAlignment.center,
          crossAxisAlignment: CrossAxisAlignment.center,
          children: [
            Text("Health is a complete harmony of the body, mind, and spirit"),
            ElevatedButton(onPressed: () => print("Starting the application..."), child: Text("Get Started"))
          ],
        ))
      )
    );


  }
}

```
And when you have multiple of such classes (i.e., screens) in your application and you need a way to navigate between them.  

This is where the Navigattor class of Flutter comes into picture.

Apps often have multiple screens- like a home page and a details page.  

Flutter uses the **Navigator** class to manage this, which acts like a stack-based controller that manages  which screen (called a route) is currently visible.  

The **Navigator** internally manages a stack of routes
- Since its a stack structure --> LIFO (Last in First Out)  

- New screens are added (pushed) to the top, and screens are removed (popped) from the top. So the last opened screen is the first one to be closed.  

- Only the topmost element on the stack (which is the current screen) is accessible to the user.
    - Example: Tapping a grocery item **pushes** the `GroceryItemScreen` onto the stack's top. When the user navigates back, that screen is **popped** off the stack.


Two core operations for any stack can also be seen in the **Navigator** which are:
- Push --> Add (navigate to) a new screen on top of the stack
    -  When you call `Navigator.push`, Flutter:
     1. Creates a new Route (e.g., a new page)
     2. Pushes it on top of the stack
     3. Displays it  

- Pop --> Remove (go back from) the current screen
    - When you call `Navigator.pop`, Flutter:
        1. Removes (pops) the topmost route
        2. Returns to the previous one

```raw


```

**`Navigator.push()`**


```dart

Navigator.push<T>(
  BuildContext context,
  Route<T> route,
)
```
- `Navigator.push()` requires two parameters:
    - `context` --> the current context (where you are now)
    - `route` --> the route to push (i.e., the page that you want to go to), typically created with `MaterialPageRoute`
        ```dart
        MaterialPageRoute<T>({
            required WidgetBuilder builder, // <== This will be our main focus
            RouteSettings? settings,
            bool maintainState = true,
            bool fullscreenDialog = false,
        })
        ```
        - `builder` --> This is the function that builds the widget for the route (the new screen to be displayed)
            - Typically used in the following manner:
                ```dart
                builder : (context) => SecondScreen(),
                ```
            - `context` --> It's a `BuildContext` object
                - Retains the information regarding the widget's location in the widget tree.
                - You cannot pass anything else.  
                
            - `SecondScreen()` --> The widget instance returned by the builder, representing the screen you want to show. This is basically your class where your widget tree for the page is located (e.g., `class LandingPage extends StatelessWidget {...}`)
```raw

```

**`Navigator.pop()`**

```dart

Navigator.pop(context)

```
- `context` --> a fixed argument that cannot be changed
    - retains the information regarding the widget's location in the widget tree.


Very simple to use, it is used to remove the current route from the navigation stack and return to the previous screen. This is how you "go back".

```raw

```

A route is essentially a screen or page in Flutter. When you navigate between screens, you are actually pushing routes on or off a stack.   

So every time you navigate between pages, you're actually manipulating routes. 

There are two main ways to define routes:
1. Implicit Routes
    - You create the route directly when navigating  

    - For instance:
        ```dart
        Navigator.push(
            context,
            MaterialPageRoute(builder: (context) => const DetailsScreen()),
        );
        ```
        - This is called implicit because you're not naming or registering the route beforehand
    ```raw

    ```
2. Explicit Routes
    - You define the routes explicitly inside your `MaterialApp` using the `routes` prameter of it.  

    - This approach is scalable for large apps
    ```dart
    MaterialApp(
    routes: {
        '/home': (context) => const HomeScreen(),
        '/details': (context) => const DetailsScreen(),
    },
    );

    ```


Implicit routes are straightforward, for instance when a certain button is pressed, you just use `Navigator.push()` or `Navigator.pop()` in the `onPressed` function. And it is as simple as that.  

Let's talk a bit more on Explicit Routes though:  
In the `MaterialApp` class, the `routes` parameter is defined as such:
```dart

MaterialApp{
    //...
    final Map<String, WidgetBuilder> routes;
}

```
- So the `routes` parametes expects a map (dictionary) where:
    - Each key is a `String` that retains the route name  
    - Each value is a function (`WidgetBuilder`) that returns a Widget (the screen to display).  

- Also note that hierarchical path-like format is used, hence the `/`

```raw

```

**`Navigator.popUntil()`**



`Navigator.popUntil` removes (pops) screens from the top of the stack until it reaches the specified route.  

It only goes back; it does not add any new screen.  

The genneral sytanx is as follows:
```dart

Navigator.popUntil(
  BuildContext context,
  RoutePredicate predicate,
);


```
`context` --> again a fixed parameter, cannot be changed, which retains the information about the location of the widget within the widget tree.  

- `predicate` --> A function that takes a `Route` objetc and returns a `bool`. It decides which route to stop at.
  - Flutter will call this function once for each route in the stack, starting from the top.
  - It keeps popping routes until the predicate returns `true`
  - When the predicate returns true, Flutter stops popping- and that route becomes the current screen.
  - `ModalRoute.withName(routePath)` is commonly used as this parameter:
    ```dart
    ModalRoute.withName(routePath)
    ```
    - `routePath` --> a string representing the named route you defined in your `MaterialApp.route`, for instance `'/homeScreen'`
```raw

```
Example:
```dart
Navigator.popUntil(context, ModalRoute.withName('/home'));


```

**`Navigator.pushAndRemoveUntil()`**

It **pushes a new screen** and then removes previous screens from the navigation stack unitl a predicate returns `true`. So:
- It **adds** a screen
- And then **deletes some or all** screens behind it   

Esenntially the same logic with `Navigator.popUntil()` with the main difference being it pushes a route on top of the stack.  


The general sytnax is as follows:
```dart

Navigator.pushAndRemoveUntil(
  BuildContext context,
  Route newRoute,
  RoutePredicate predicate,
);

```
- The only new paramaeter is `newRoute`, the rest of the parameters work and used in the same manner as `Navigator.popUntil()`
    - `newRoute` --> the new screen (route) you want to push
      - Most commonly a `MaterialPageRoute` *(same usage as in `Navigator.push()`)*


```raw

```

**`initalRoute`**

This property specifies the first route (or screen) that the app should display when it starts.  

When the app launches, Flutter looks at the `initalRoute` value and matches it with a route defined in the routes map.  

It is part of the `MaterialApp` configuration and overrides the `home: ` parameter.
- If `initialRoute` is provided:
    - Flutter ignores the `home: ` widget
    - Instead it goes to the route whose name matches `initalRoute` in the `routes: ` map.
- If `initialRoute` is not provided:
    - Flutter uses the widget defined in `home: ` as the inital route.
```raw

```
Example use case:
```dart
MaterialApp(
  initialRoute: '/home',
  routes: {
    '/home': (context) => HomeScreen(),
    '/details': (context) => DetailsScreen(),
    '/settings': (context) => SettingsScreen(),
  },
);

```
❗ Note that `initalRoute` must match a key of the map provided in `routes`.  

`initialRoute` allows you to conditionally decide upon which page to be the landing page. For instance:
```dart
initialRoute: userLoggedIn ? '/dashboard' : '/login'
```



### IMPORTANT !!!!!

In Flutter, the `context` you pass to `Navigator` methods must come from a location in the widget tree where a `Navigator` already exists above it. Because the Navigator is provided by widgets like `MaterialApp` or `CupertinoApp`, any attempt to call navigation functions using a context that originates above those widgets—or from a subtree that is not wrapped by them—will fail.  



That’s why we typically call` runApp(MaterialApp(home: HomePage()))`: doing so ensures that every context used throughout the app originates from within the same widget tree whose root includes a `MaterialApp`. Since `MaterialApp` is what provides the `Navigator`, wrapping the entire app under it guarantees that all descendant widgets share a common ancestor that contains a `Navigator`. As a result, any context passed to navigation methods will correctly resolve a Navigator above it, preventing errors and keeping the widget tree structurally consistent.

```raw

```
See the example below:

```dart

import 'package:flutter/material.dart';

void main() {
  return runApp(MaterialApp(
    home: HomePage()                  // <<< ================================================ SEE HERE
  ));
}

//////////////////////////// FIRST PAGE ////////////////////////////

class HomePage extends StatelessWidget {

  const HomePage({super.key});


  @override
  Widget build(BuildContext context){
    return Scaffold(
        appBar: AppBar(
          title: Text("CounterApp"),
        ),
      body: Center(child:Column(
        mainAxisAlignment: MainAxisAlignment.center,
        crossAxisAlignment: CrossAxisAlignment.center,
        children: [
          Text("Welcome to the Counter App!"),
          ElevatedButton(onPressed: () => Navigator.push(context, MaterialPageRoute(builder: (context) => CounterPage())), // <============ HERE
              child: Text("Proceed to the Counter Page"))
        ],
      ),),
      );
  }
}


//////////////////////////// SECOND PAGE ////////////////////////////

class CounterPage extends StatefulWidget {
  const CounterPage({super.key});

  @override
  State<CounterPage> createState() => _StateCounterPage();
}

class _StateCounterPage extends State<CounterPage> {

  int _cntr = 0;

  void _incrementCntr() {
    setState(() {
      _cntr++;
    });
  }

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(
        actions: [
          Row(
            children: [
              ElevatedButton(onPressed: () => Navigator.push(context, MaterialPageRoute(builder: (context) => HomePage())), // <============ HERE
                  child: Text("Back to Home"))
            ]
          )
        ]
      ),
    body: Center(child: Text("Counter: $_cntr")),
    floatingActionButton: FloatingActionButton(onPressed: _incrementCntr, child: Text("+"),),
    );
  }
}







## Handling User Inputs in Flutter

There are two main ways to Handle Input depending on the complexity and structure required:
1. Using the basic `TextField` for quick single inputs and manual validation.
2. Using `TextFormField` within a `Form` widget for multi-field inputs and built-in validation.

### `TextField` Widget

A fundamental widget in FLutter that is used to cpature single-line or multi-line input from the user. 

It:
- allows user to enter text
- can be styled using `InputDecoration`
- works with a `TextEditingController` to read, update, or clear text.
- does **NOT** have built-in validation, hence all validation logic, if need be, needs to be manually implemented by the coder.  

- The basic syntax is as follows:
```dart
TextField(
  TextEditingController? controller,
  InputDecoration? decoration,
)

```
- `controller` is an instance of the `TextEditingController` that stores and manages all the text input entered into the `TextField`. The instance has 3 attributes:

    3. `composing` --> IME composition (used for non-Latin languages)


- `InputDecoration()` have a lot of parameters, the most essential ones are:

    InputDecoration(

        String? hintText,
        border: OutlineInputBorder(),
        Widget? suffix, // <=== Icon or button at the end of the box
    )
    ```
    - `OutlineInputBorder()` --> gives it a visible frame

### `TextFormField` Widget

A `Form` widget in Flutter is a special widget designed to group, managae, validate, and save multiple input fields.  

It is the core structure used for:
- Login screens
- Registiration forms
- Checkout pages
- Multi-field data entry
- Contac/feedback forms  

`Form` is an important widget, as it has built-in input handling mechanisms:
- **Data Accuracy:** Ensures the user enters valid and structured data
- **User Experience:** Provides automatic error messages and guides the user.
- **Security:** Prevents invalid/malformed submissions and reduces risk.
- **Form Validation:** Allows validating all fields at once.  



The Form widget is a container that coordinates mutliple `TextFormField` widgets inside it. **So you always put `TextFormField` inside a `Form` widget.**  

The barebones general sytanx for the `Form` widget is as follows:
```dart

const Form({
  Key? key,
  required Widget child,
})

```
- `key` --> Explained later on below  

- `child` --> Usually a `Column` or `Row` containing `TextFormField`s.


```raw

```

The general syntax for `TextFormField` is as follows:
```dart

const TextFormField({
  TextEditingController? controller,
  InputDecoration decoration = const InputDecoration(),
  FormFieldValidator<String>? validator,
  FormFieldSetter<String>? onSaved,
})

```

The widget being `const` allows you to create compile-time constan widgets whenever their parameters are also constant, which is a great optimization.  

- `controller` and `decoration` --> same function and usage with the controller parameter of `TextField` widget  

- `validator` --> a callback function that defines the validation logic for its field. But has a few caveats that is:
    - If a `String` is returned, then the validation is failed and that string is the error message  

    - If `null` is returned then the validation is passed.  

    - The `validator` takes a single input that `value` that represent the current text value in the field. (same logic with `onSaved`)  
    

- `onSaved` --> a callback function used to store the final, validated value of a form field when `FormState.save()` is called.
    - It triggers only when the entire form is saved, typically inside a submit button.
    - Inside `onSaved`, you should assign the final validated value of the text field to some variable or data model.  
    - For instance:
        ```dart

        onSaved: (value) => _username = value,
        ```
        - **When .save() is called, the `value` passed to the callback function is the current text value in the field. Hence your callback function requires a parameter that represents the current text value in the field**
```raw

```

As discussed previously, the Stateful Widgets themselves are stateless and what allows them to Stateful is a seperate companion object called the `State`.  

For the case of the `Form` widget, the state is in a `FormState` object hidden inside the Form.  

To acceess this FormState, you need something that points exactly to that specific form. **That "something" is a Global Key**.  

Each widget has a `Key` and a key acts like an ID card for the widget. It helps flutter to:
- Identify
- Differentiate
- and access a specific widget anywhere in the tree.  

A `GlobalKey` is a specific type of `Key` that is visicle from anywhere, not only inside the widget. It gives you:
- a reference to the widget
- AND a reference to that widget's internal state object.  

This means that, if you initialize a GlobalKey for your `Form` and attach it with your `Form`, then by using that `GlobalKey` anywhere within the code you can perform certain operation for all `TextFormField` widgets defined within that `Form`.  

First things first, you initialize a GlobalKey and attach it with your form widget in the following manner:
```dart

// Initializing the Global FormKey
GlobalKey<FormState> _formKey = GlobalKey<FormState>();

// Binding the Global FormKey with the Form Widget:
Form(
    key: _formKey,
    child: ...
)

```
Then, anywhere within your code, you can write:
```dart

_formKey.currentState!.validate();
_formKey.currentState!.save();
_formKey.currentState!.reset();

```
- `.validate()` --> Run ALL validators of every `TextFormField` within that `Form`. 
    - If all validators return `null` (no errors), then `.validate()` returns `true`.

- `save()` --> Call every `onSaved()` callback  

- `reset()` --> Clear all fields & errors  

- `currentState` --> it is a property of the `GlobalKey` and when you call `_formKey.currentState` it returns the state object of the widget attached to that key
    - Since `curenntState` may be `null` and we can't call `.validate()` and other methods for nullable types, we use null-assertion operator (`!`).
    - Otherwise we would get a compiler error.

```raw

```

Logically you first call `.validate()` and then call `.save()`

**`ListView.builder`**

The `ListView.builder` is a Flutter widget used for efficiently rendering dynamic lists that are built from data structures like a Dart `List`.  

It is used when the list is dynamic, long, or coming from data structures like `List<Model>`.  

It builds widgets on demand (lazily). It only creates widgets necessary to display the items currently visible on the screen, plus a few nearby items, making it suitable for very large or potentially infinite lists.  

Note that `ListView,builder` works on any data source as long as you are able to proved `itemCount` and `itemBuilder`. Thw widget itself has no requitement that the data must be a `List`.  


The general sytnax is as follows:
```dart

ListView.builder({
  required int itemCount,
  required Widget Function(BuildContext context, int index) itemBuilder,

})

```
- `itemCount` --> Total number of items in the list.  

- `itemBuilder` --> A function that returns one widget per index.
    - The function should take 2 parameters, which are `context` (for navigating to the widget tree) and `index` which is the index of the current element of the list. These are automatically passed on by the Flutter.  

    - The function should return a widget, which is usually `ListTile`.  

```dart

ListView.builder(
  itemCount: students.length,
  itemBuilder: (context, index) {
    final s = students[index];
    return ListTile(
      title: Text(s.name),
      subtitle: Text(s.id),
    );
  },
)

```

As for the general syntax of `ListTile`, it is:
```dart

ListTile({
  Widget? leading,
  Widget? title,
  Widget? subtitle,
  Widget? trailing,
  GestureTapCallback? onTap,

})

```
- `leading` --> A widget displayed at the start of the (left side)`ListTile`
    - Icons are commonly used
    - Can be any widget 

- `title` --> The primary text of the ListTile
    - Can be any widget, usually `Text(...)`  

- `subtitle` --> A secondary, smaller text below the title.
    - Used for extra info
    - Usually `Text(...)`  

- `trailing` --> A widget displayed at the end of the `ListTile` (right side)
